```python
# RIFAH TASFIA
# Tic-Tac-Toe Game
import random
# Initialize the game board
board = ["-", "-", "-",
        "-", "-", "-",
        "-", "-", "-"]
# Initialize current player
currentPlayer = "X"

# Initialize winner
winner = None

# Initialize game state
gameRunning = True

# game board for Tic-Tac-Toe
# I printed the elements of the board list in a formatted grid.
def printBoard(board):
    print(board[0] + " | " + board[1] + " | " + board[2])
    print("---------")
    print(board[3] + " | " + board[4] + " | " + board[5])
    print("---------")
    print(board[6] + " | " + board[7] + " | " + board[8])


# Function to take player input
# This function will handle player moves in the Tic-Tac-Toe game
def playerInput(board):
    inp = int(input("Select a spot 1-9: "))
    if board[inp-1] == "-":
        board[inp-1] = currentPlayer
    else:
        print("Oops player is already at that spot.")


# check for win or tie
# Examines each row of the board to see if any player has filled all three spots with their marker("X" or "O")
# It updates the winner variable if a win is detected.
def checkHorizontal(board):
    global winner
    if board[0] == board[1] == board[2] and board[0] != "-":
        winner = board[0]
        return True
    elif board[3] == board[4] == board[5] and board[3] != "-":
        winner = board[3]
        return True
    elif board[6] == board[7] == board[8] and board[6] != "-":
        winner = board[6]
        return True

# Function to check for a win in each row of the board
# It examines each row to see if any player has filled all three spots with their marker ("X" or "O")
def checkRow(board):
    global winner
    if board[0] == board[3] == board[6] and board[0] != "-":
        winner = board[0]
        return True
    elif board[1] == board[4] == board[7] and board[1] != "-":
        winner = board[1]
        return True
    elif board[2] == board[5] == board[8] and board[2] != "-":
        winner = board[2]
        return True

# Function to check for diagonal win
def checkDiag(board):
    global winner
    if board[0] == board[4] == board[8] and board[0] != "-":
        winner = board[0]
        return True
    elif board[2] == board[4] == board[6] and board[4] != "-":
        winner = board[2]
        return True

# Function to check if there's a winner
def checkIfWin(board):
    global gameRunning
    if checkHorizontal(board):
        printBoard(board)
        print(f"The winner is {winner}!")
        gameRunning = False

    elif checkRow(board):
        printBoard(board)
        print(f"The winner is {winner}!")
        gameRunning = False

    elif checkDiag(board):
        printBoard(board)
        print(f"The winner is {winner}!")
        gameRunning = False

# Function to check if it's a tie
def checkIfTie(board):
    global gameRunning
    if "-" not in board:
        printBoard(board)
        print("It is a tie!")
        gameRunning = False


# Function to switch player
def switchPlayer():
    global currentPlayer
    if currentPlayer == "X":
        currentPlayer = "O"
    else:
        currentPlayer = "X"

# Function for computer's move
def computer(board):
    while currentPlayer == "O":
        position = random.randint(0, 8)
        if board[position] == "-":
            board[position] = "O"
            switchPlayer()

# Main game loop
# Loop continues until the game ends
while gameRunning:
    printBoard(board)  #Display the current state of the board
    playerInput(board) #Allow the player to make a move
    checkIfWin(board) #Check if there is a winner after each player move
    checkIfTie(board) #Check if the game ends in a tie after each player move
    switchPlayer()  #Switch to the next player's turn
    computer(board) #Let the computer make a move
    checkIfWin(board) #Check if there is a winner after the computer's move
    checkIfTie(board) #Check if the game ends in a tie after the computer's move
```
